```python
## Load libraries
import pandas as pd
import numpy as np
```

```python
#csv file name to be read in 


test_data = pd.read_csv("D:/RStudio/GreatUrbanShift/data/GBIFdatabase/0270981-200613084148143.csv", nrows=5, header=0, sep="\t") 

### Preview what the data looks like
test_data.head()
list(test_data)[30]

```

```python
### Identify the columns to keep

usecols = list(test_data) 
usecols

keepCols = ["gbifID","datasetKey","phylum","class","order","family","genus","species","day","month","year","decimalLatitude","decimalLongitude"]
dataSelect = test_data[keepCols]
dataSelect.head()
```

```python
### Split into separate CSV files

## total number of observations
number_lines = 547547455  

#size of rows of data to write to the csv, 
rowsize = 1000000 

## Columns to keep
keepCols = ["gbifID","datasetKey","phylum","class","order","family","genus","species","day","month","year","decimalLatitude","decimalLongitude"]
    
#start looping through data writing it to a new file for each set

for i in range(1,number_lines,rowsize):
    df = pd.read_csv("D:/RStudio/GreatUrbanShift/data/GBIFdatabase/0270981-200613084148143.csv", sep="\t",
          header=0,
          nrows = rowsize,#number of rows to read at each loop
          skiprows = i)#skip rows that have been read

    ## select only core columns
    dfcore = df[keepCols]
    
    #csv to write data to a new file with indexed name. input_1.csv etc.
    out_csv = 'D:/RStudio/GreatUrbanShift/data/GBIFdatabase/split/input' + str(i) + '.csv'
    dfcore.to_csv(out_csv,
          index=False,
          header=0,
          mode='w',#append data to csv file
          chunksize=rowsize)#size of data to append for each loop
```
